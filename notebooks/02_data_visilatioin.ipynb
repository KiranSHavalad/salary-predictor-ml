# Data Visualization

## Project: Smart Employee Salary Predictor

### Objective

In this notebook, we will visualize the dataset to understand:

- Salary distribution
- Experience vs Salary
- Remote work trends
- Company size distribution
- Most common job titles
- Outliers

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/raw/salaries.csv")
```

```python
plt.figure(figsize=(10,6))

plt.hist(df["salary_in_usd"], bins=30)

plt.title("Salary Distribution")
plt.xlabel("Salary (USD)")
plt.ylabel("Number of Employees")

plt.show()
```

```python
df["experience_level"].value_counts().plot(kind="bar")

plt.title("Employees by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Count")

plt.show()
```

```python
df.groupby("experience_level")["salary_in_usd"].mean().plot(kind="bar")

plt.title("Average Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary (USD)")

plt.show()
```

```python
df["remote_ratio"].value_counts().plot(kind="bar")

plt.title("Remote Work Distribution")
plt.xlabel("Remote Ratio")
plt.ylabel("Employees")

plt.show()
```

```python
df["company_size"].value_counts().plot(kind="bar")

plt.title("Company Size Distribution")
plt.xlabel("Company Size")
plt.ylabel("Count")

plt.show()
```

```python
df["job_title"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(df["salary_in_usd"])

plt.title("Salary Outliers")

plt.ylabel("Salary")

plt.show()
```

```python
df.groupby("company_size")["salary_in_usd"].mean().plot(kind="bar")

plt.title("Average Salary by Company Size")
plt.xlabel("Company Size")
plt.ylabel("Average Salary")

plt.show()
```

# Conclusion

From today's visualizations, we learned:

- The salary distribution is not perfectly uniform.
- Experience level appears to influence salary.
- Company size may also affect salary.
- Remote work distribution provides insight into work arrangements.
- Some salaries are potential outliers.
- Certain job titles are much more common than others.

These observations will help us decide how to preprocess the data before training our model.
